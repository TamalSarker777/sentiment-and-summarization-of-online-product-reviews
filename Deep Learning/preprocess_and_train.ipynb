```python
#removing emojis from dataset
import re
def remove_emojis(text):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', text)

count=0
for reviews in df['reviews']:
    df.iloc[count, 1]=remove_emojis(reviews)
    count=count+1
```

```python
def remove_english(text):
    bangla_characters = "ঁংঃঅআইঈউঊঋএঐওঔকখগঘঙচছজঝঞটঠডঢণতথদধনপফবভমযরলশষসহড়ঢ়য়০১২৩৪৫৬৭৮৯া-ৃৄৢৣ ্রেৈোৌিীুূ"
    bangla_text = ""
    for char in text:
        if char in bangla_characters:
            bangla_text += char
    return bangla_text

count=0
for reviews in df['reviews']:
    df.iloc[count, 1]=remove_english(reviews)
    count=count+1
```

```python
def convert_number_to_words(number):
    words = {
        0: "শূন্য",
        1: "এক",
        2: "দুই",
        3: "তিন",
        4: "চার",
        5: "পাঁচ",
        6: "ছয়",
        7: "সাত",
        8: "আট",
        9: "নয়",
        10: "দশ",
        20: "বিশ",
        30: "তেরে",
        40: "চৌদ্দ",
        50: "পঞ্চাশ",
        60: "ষাট",
        70: "সতের",
        80: "আঠাশ",
        90: "নব্বই",
        100: "একশ",
        1000: "হাজার"
    }


    if number in words:
        return words[number]

    if number < 100:
            return words[10 * (number//10)] + words[number%10]

    if number < 1000:
            return words[number//100] + "শ" + " " + convert_number_to_words(number % 100)

    if number < 5000:
            return convert_number_to_words(number // 1000) + " " + words[1000] + " " + convert_number_to_words(number % 1000)


    return





# # number = int(input("Enter a number to convert to words (1-5000): "))
# # print(convert_number_to_words(number))
```

```python
count=0
for reviews in df['reviews']:
#     df.iloc[count, 0]=remove_english(reviews)
#     count=count+1

    words = reviews.split() #split the words

    temp=[]
    for word in words:
        try:
            int(word)
            temp.append(convert_number_to_words(int(word)))
        except ValueError:
            temp.append(word)

    #converting list to string
    mystring=' '
    for x in temp:
        try:
            mystring += ' '+x
        except:
            continue


#input the modified sentences to the dataset
    df.iloc[count, 1] = mystring
    count=count+1
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
from indicnlp.tokenize import indic_tokenize
```

```python
X =  df['reviews'].values
Y = df['sentiments'].values
product_names = df['ProductName'].values
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(Y)
y
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test, product_train, product_test = train_test_split(
    X, y, product_names, test_size=0.2, random_state=42
)
```

```python
# Tokenize Bengali text
def tokenize_bengali_text(text):
    tokens = indic_tokenize.trivial_tokenize(text)
    return ' '.join(tokens)
```

```python
X_train = [tokenize_bengali_text(text) for text in X_train]
X_test = [tokenize_bengali_text(text) for text in X_test]
```

```python
# Tokenize the text data
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)
```

```python
tokenizer
```

```python
# Convert text to sequences
X_train_sequences = tokenizer.texts_to_sequences(X_train)
X_test_sequences = tokenizer.texts_to_sequences(X_test)
```

```python
X_train_sequences
```

```python
# Pad sequences for uniform length
X_train_padded = pad_sequences(X_train_sequences, maxlen=100, truncating='post', padding='post')
X_test_padded = pad_sequences(X_test_sequences, maxlen=100, truncating='post', padding='post')
```

```python
type(X_train_padded)
```

```python
# Build the model
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=16, input_length=100))
model.add(Bidirectional(LSTM(64, return_sequences=True)))
model.add(layers.Dropout(0.5))  # Add dropout layer
model.add(Bidirectional(LSTM(64)))
model.add(layers.Dropout(0.5))  # Add dropout layer
model.add(Dense(3, activation='softmax'))
```

```python
# Compile the model
model.compile(
    loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']
    )
```

```python
# Add EarlyStopping callback with provided configuration
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
```

```python
# Train the model
model.fit(X_train_padded,
          y_train, epochs=5,
          batch_size=40,
          validation_data=(X_test_padded, y_test),
          callbacks=[early_stopping]
          )
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test_padded, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
import numpy as np
from tensorflow.keras.models import load_model

# Load the saved model
loaded_model = load_model("sentiment_model.h5")

# Define a mapping dictionary
sentiment_mapping = {0: 'negative', 1: 'neutral', 2: 'positive'}

# Function to predict sentiment
def predict_sentiment(input_text):
    # Tokenize Bengali text
    tokens = indic_tokenize.trivial_tokenize(input_text)
    tokenized_text = ' '.join(tokens)

    # Tokenize the text data
    sequences = tokenizer.texts_to_sequences([tokenized_text])

    # Pad sequences for uniform length
    padded_sequence = pad_sequences(sequences, maxlen=100, truncating='post', padding='post')

    # Make prediction using the loaded model
    predictions = loaded_model.predict(padded_sequence)

    # Decode the predicted sentiment label
    predicted_label = np.argmax(predictions)
    predicted_sentiment = sentiment_mapping[predicted_label]

    return predicted_sentiment

# Example usage
user_input = input("Enter a Bengali review text: ")
predicted_sentiment = predict_sentiment(user_input)
print(f"Predicted Sentiment: {predicted_sentiment}")
```
